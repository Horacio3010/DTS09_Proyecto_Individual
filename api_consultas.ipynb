Película (sólo película, no serie, ni documentales, etc) con mayor duración según año, plataforma y tipo de duración. La función debe llamarse get_max_duration(year, platform, duration_type) y debe devolver sólo el string del nombre de la película.


```python
def get_max_duration(year: int, platform: str, duration_type: str) -> str:
    # Filtrar sólo las películas (y no series, documentales, etc) del año y plataforma especificados
    df_filtered = dfcatalogo[(dfcatalogo["type"] == "Movie") & (dfcatalogo["release_year"] == year) & (dfcatalogo["listed_in"].str.contains(platform, case=False))]
    
    # Si no hay películas que coincidan, regresar None
    if df_filtered.empty:
        return None
    
    # Extraer las duraciones y los nombres de la película
    df_filtered["duration"] = df_filtered["duration"].str.extract("(\d+)").astype(int)
    durations = df_filtered["duration"].tolist()
    names = df_filtered["title"].tolist()
    
    # En función del tipo de duración, encontrar el índice de la película con la duración máxima
    if duration_type == "min":
        max_idx = durations.index(max(durations))
    elif duration_type == "season":
        max_idx = durations.index(max(durations)) # aquí podrás profundizar en la columna "duration_type"
    else:
        # Si un tipo de duración no válido se especifica, regresar None
        return None
    
    # Regresar el nombre de la película con la duración máxima
    return names[max_idx]
```

```python
movie_name = get_max_duration(2012, "netflix", "min")
print(movie_name)
```

Cantidad de películas (sólo películas, no series, ni documentales, etc) según plataforma, con un puntaje mayor a XX en determinado año. La función debe llamarse get_score_count(platform, scored, year) y debe devolver un int, con el total de películas que cumplen lo solicitado.


```python
def get_score_count(platform, scored, year):
    # Select rows where type is "movie" and the platform matches
    df_platform = dfcatalogo[(dfcatalogo['type'] == 'movie') & (dfcatalogo['plataforma'] == platform)]

    # Filter by release year and rating
    df_filtered = df_platform[(df_platform['release_year'] == year) & (df_platform['avg_rating'] >= scored)]

    # Return the row count as an integer
    return int(df_filtered.shape[0])
```

```python
# Buscar películas con puntaje mayor a X.0 en X plataforma en X año
count = get_score_count('netflix', 3.0, 2012)

# Imprimir el resultado
print(f"La plataforma  lanzó {count} películas con un puntaje mayor a 3.0 en 2012")
```

Actor que más se repite según plataforma y año. La función debe llamarse get_actor(platform, year) y debe devolver sólo el string con el nombre del actor que más se repite según la plataforma y el año dado.


```python
def get_actor(platform, year):
    # Filtremos el dataframe dfcatalogo por la plataforma y el año dados
    filtros = (dfcatalogo['plataforma'] == platform) & (dfcatalogo['release_year'] == year)
    
    # Creeamos un dataframe que contenga solo la columna de 'cast' con los datos filtrados 
    actores = pd.DataFrame(dfcatalogo.loc[filtros, 'cast'])
    
    # Explotamos la lista de actores en una columna separada.
    actores = actores['cast'].str.split(', ', expand=True)
    
    # Obtenemos el recuento de cada actor y los sumamos por columna
    actores_count = actores.apply(pd.Series.value_counts).sum(axis=1).reset_index()
    
    # Renombramos las columnas del dataframe
    actores_count.columns = ['Actor', 'Count']
    
    # Ordenamos el dataframe en orden descendente según la cantidad de veces que aparece cada actor
    actores_count = actores_count.sort_values('Count', ascending=False)
    
    # Devolvemos el nombre del actor que más aparece
    return actores_count.iloc[0][0]
```

```python
# Encontrar el actor más común en Netflix en 2017
actor = get_actor('netflix', 2017)

# Imprimir el resultado
print(f"El actor más común de la plataforma en 2017 fue {actor}")
```

La cantidad de contenidos/productos (todo lo disponible en streaming) que se publicó por país y año. 
La función debe llamarse prod_per_county(tipo,pais,anio) deberia devolver el tipo de contenido (pelicula,serie,documental) por pais y año en un diccionario con las variables llamadas 'pais' (nombre del pais), 'anio' (año), 'pelicula' (tipo de contenido).


```python
def prod_per_county(tipo, pais, anio):
    # Filtremos el dataframe dfcatalogo por el tipo de contenido, país y año dados
    filtros = (dfcatalogo['type'] == tipo) & (dfcatalogo['country'] == pais) & (dfcatalogo['release_year'] == anio)
    
    # Contemos la cantidad de productos que cumplen con los criterios de filtro
    count = dfcatalogo.loc[filtros].shape[0]
    
    # Creamos un diccionario con los resultados
    resultado = {'pais': pais, 'anio': anio, tipo: count}
    
    # Devolvemos el diccionario
    return resultado
```

La cantidad total de contenidos/productos (todo lo disponible en streaming, series, documentales, peliculas, etc) 
según el rating de audiencia dado (para que publico fue clasificada la pelicula). La función debe llamarse get_contents(rating) 
y debe devolver el numero total de contenido con ese rating de audiencias.


```python
def get_contents(rating):
    # Filtremos el dataframe dfcatalogo por el rating de audiencia dado
    filtros = dfcatalogo['rating'] == rating
    
    # Calculemos la cantidad total de contenidos/productos con ese rating
    cantidad = dfcatalogo.loc[filtros].shape[0]
    
    # Devolvemos la cantidad total de contenidos/productos con ese rating
    return cantidad
```
