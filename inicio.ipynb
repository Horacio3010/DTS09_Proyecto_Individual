Primera Consigna:
Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets 
(ejemplo para títulos de Amazon = as123)


```python
# Definir función para generar el campo id
def generar_id(row):
    plataforma = 'Amazon'
    show_id = row['show_id']
    primera_letra = plataforma[0].lower()
    
    return f"{primera_letra}{show_id}"

# Aplicar la función a cada fila y crear la nueva columna 'id'
amazondf['id'] = amazondf.apply(generar_id, axis=1)

# Verificar que la nueva columna 'id' se haya creado correctamente
print(amazondf.head())
```

```python
# Cambiar el nombre de la columna 'id' a 'movieId'
amazondf.rename(columns={'id': 'movieId'}, inplace=True)

# Verificar que la operación se haya hecho correctamente
print(amazondf.columns)
```

Concatenar los dataframes para poder trabajar mejor los datos

```python
# Agregar una columna 'plataforma' a cada DataFrame para identificar la plataforma
amazondf['plataforma'] = 'Amazon'
disneydf['plataforma'] = 'Disney+'
huludf['plataforma'] = 'Hulu'
netflixdf['plataforma'] = 'Netflix'

# Concatenar los DataFrames en uno solo
dataframes = [amazondf, disneydf, huludf, netflixdf]
df_catalogo = pd.concat(dataframes, ignore_index=True)

# Verificar que los DataFrames se hayan concatenado correctamente
df_catalogo.head()
```

```python
# crear la columna movieId concatenando la inicial de la plataforma y el show_id
df_catalogo['movieId'] = df_catalogo.apply(
    lambda row: row['plataforma'][0].lower() + str(row['show_id']),
    axis=1
)

# guardar el resultado en un nuevo archivo CSV
df_catalogo.to_csv('df_catalogo_with_movieId.csv', index=False)

# Verificar que la columna 'id' se haya eliminado correctamente
print(df_catalogo.head())
```

Segunda Consigna: Los valores nulos del campo rating deberán reemplazarse por el string “G” corresponde al maturity rating: “general for all audiences”

```python
# Reemplazar valores nulos en la columna 'rating' con el string 'G'
df_catalogo['rating'].fillna('G', inplace=True)

# Verificar que la operación se haya hecho correctamente
print(df_catalogo['rating'].unique())
```

Tercera Consigna: De haber fechas, deberán tener el formato AAAA-mm-dd


```python

df_catalogo['date_added'] = pd.to_datetime(df_catalogo['date_added'], errors='coerce')
```

Cuarta Consigna: Los campos de texto deberán estar en minúsculas, sin excepciones


```python

df_catalogo = df_catalogo.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
```

```python
df_catalogo.sample(10)
```

Quinta Consigna: El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)


```python

df_catalogo[['duration_int', 'duration_type']] = df_catalogo['duration'].str.extract('(\d+) (\w+)')

# convertir duration_int a tipo numérico
df_catalogo['duration_int'] = pd.to_numeric(df_catalogo['duration_int'])


# guardar el resultado en un nuevo archivo CSV
df_catalogo.to_csv('df_catalogo_v1.csv', index=False)
```
